```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("log.json") as dataFile:
    data = json.load(dataFile)
    for key,value in data.items():
        df = pd.DataFrame(value)

df['http-version'] = pd.to_numeric(df['http-version'])
df['ResponseTimein ms'] = pd.to_numeric(df['ResponseTimein ms'])
df['date'] = pd.to_datetime(df['date'])
# df.info()
# df.head(10)
# df.dtypes

```

```python
fig, axs = plt.subplots()
axs.scatter(df['Method'], df['ResponseTimein ms'])
axs.set_xlabel("Method")
axs.set_ylabel("Response Time ms")
axs.set_title("Api Response Time")
plt.show()
```

```python
fig, axs = plt.subplots()
df.groupby(df.date.dt.hour).apicall.count().plot(ax=axs,xlabel='Hour',ylabel='No. of Requests',title='Api Peak Hour',style="k--")
```
